# Experiment 1: R = 10000 Ohms

```python
# Import sensor data
data = np.genfromtxt('Sensor_Values_R1_10k_V2.txt', delimiter=',')

# Plot sensor data
time_step = 0.01 # 10ms
time = np.arange(0, len(data)*time_step, time_step)

plt.plot(time, data, label='Sensor 1')
plt.xlabel('Time (s)')
plt.ylabel('Sensor Value (mV)')
plt.title('Sensor Data for R1 = 10kOhm')

```

# Comments on differences between the two experiments: 

First, there is an error in the question, the resistors are in SERIES with the photoresistor, not in parallel as stated in the question.

Basically, the resistor acts as a voltage divider. Letting the resistance of the resistor be $R$ and the resistance of the photoresistor be $R_{photo}$, the voltage read by the Arduino is given by:

$$V_{read} = V_{in} \frac{R}{R + R_{photo}}$$
In experiment 1, the the resistance $R = 10 $ k ohms is much larger than in experiment 2 where $R = 560$ ohms. This means that the voltage read by the Arduino is much closer to $V_{in}$ in experiment 1 than in experiment 2. This is why the voltage read by the Arduino is much higher in experiment 1 than in experiment 2.

It should be noted that the reading displayed in the graph is the un calibrated reading of the photoresistor and the arduino. THe actual signal going to the piezo is calibrated to be between 0 and 5V to take into account the ambient lighting. This is why the piezo is not as loud in experiment 1 as in experiment 2.
